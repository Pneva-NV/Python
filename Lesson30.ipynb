```python
# Создать новый столбец в таблице с понгвинами, который будет отвечать за показатель длины клюва пингвина.
# high - длинный (42)
# middle - средний (от 35 до 42)
# low - маленький (до 35)
```

```python
new_penguins["bill_length_mm_new"] 
```

```python
new_penguins.loc[(new_penguins['bill_length_mm'] >= 42), 'bill_length_mm_new'] = 'high'
new_penguins.loc[((new_penguins['bill_length_mm'] >= 35) & (new_penguins['bill_length_mm'] < 42)), 'bill_length_mm_new'] = 'middle'
new_penguins.loc[(new_penguins['bill_length_mm'] < 35), 'bill_length_mm_new'] = 'low'
```

```python
new_penguins
```

```python
new_penguins.rename({"bill_length_mm_new":"height_group"}, axis=1, inplace=True) #Переименование столбца
```
